# Defining Function to fetch the articles

```python
def fetch_Articles(url):
    Articles = []
    data = requests.get(url)
    soup = BeautifulSoup(data.content,'html.parser')
    for i in soup.find_all('h3', class_=['entry-title td-module-title']):
        Articles.append(i.find('a')['title'])
    return Articles
```

```python
Articles = fetch_Articles(url)
```

```python
#We will fetch the data from various pages of the website
```

```python
url_list = []
for i in range(2,21,1):
    url = "https://www.opindia.com/latest-news/page/" + str(i) + "/"
    url_list.append(url)
```

```python
url_list
```

# Fetching data from all these urls

```python
all_articles = []
for i in url_list:
    all_articles.extend(fetch_Articles(i))
```

```python
len(all_articles)
```

```python
all_articles = list(set(all_articles))
```

```python
all_articles
```

# Tokenizing and Stemming the data

```python
p_art = []
for i in all_articles:
    q = i.upper()
    from re import sub
    q = sub("[^A-Z,a-z,0-9 ]","",q)
    from nltk.stem import PorterStemmer
    tk_q = q.split(" ")
    sent = ""
    for j in tk_q:
        ps = PorterStemmer()
        sent = sent + " " + ps.stem(j).upper()
    p_art.append(sent)
```

```python
p_art
```

# Vectorizing the Articles

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer()
A = tf.fit_transform(p_art).toarray()
```

# Clustering into 5 groups

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=5)
model = km.fit(A)
model.labels_
```

# Creating the DataFrame

```python
import pandas as pd
B = pd.DataFrame(all_articles,columns=['Articles'])
```

```python
B['Cluster'] = model.labels_
```
